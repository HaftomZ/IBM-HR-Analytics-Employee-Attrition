##   - Data Preparation and Cleaning:

```python
na_count = df.isna().sum().sum()
print("Number of NA values in the data frame: ", na_count)
```

```python
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows: ", duplicate_count)
```

```python
### Convert categorical string columns into numeric codes :
#-----------------------------------------------------------
# Get categorical variable columns unique values series
df1 = df.loc[:, df.dtypes == 'object']
string_value_list= pd.Series({c: df1[c].unique() for c in df1})
cat_cols = string_value_list.keys()
print(string_value_list)

# Replace string data with their index value in the "string_value_list" variable
for col, uni_val in string_value_list.items():
    map_dict = {k: v for v, k in enumerate(uni_val)}
    df[col]= df[col].map(map_dict)
#-----------------------------------------------------------
```

# 1. What is the attrition rate for the company, and what are the factors that contribute to it?

```python
plt.hist(df['Attrition'], bins = 10)
plt.show()
```

```python
#Ratios
print('The percentage of the people who leave the company: %', round(count1/(count1+count2)*100))
```

# 6. How the Ages affects the Attrition rate

```python
plt.hist(df['Age'], bins = 10)
plt.show()
```

```python
#Average age of the personal who left the company
average_age_yes = df[df['Attrition'] == 0]['Age'].mean()
print("Average Age for Attrition:", round(average_age_yes,2))
```

```python
#Average age of the personal who left the company
average_age_no = df[df['Attrition'] == 1]['Age'].mean()
print("Average Age for Employees:", round(average_age_no, 2))
```

```python
average_age = df.groupby('Attrition')['Age'].mean()
average_age.plot(kind='bar', color='blue')
plt.xlabel('Attrition')
plt.ylabel('Average Age')
plt.title('Average Age by Attrition')
```
